```python
def plot_fn(rewards, avg_rewards):
    fig = plt.figure(figsize=(10, 7))
    gs = fig.add_gridspec(3, 7)
    ax = fig.add_subplot(gs[:3, :3])
    ax2 = fig.add_subplot(gs[:3, 4:7])
    
    ax.cla()
    print(ax)
    ax.set_title(f'test_rewards')
    ax.plot(rewards)
    
    ax2.cla()
    ax2.set_title(f'training_rewards')
    ax2.plot(avg_rewards)
    plt.show()
    
#plot_fn([i*2 for i in range(100)], [((i*2) + (i/2) * 3) for i in range(100)])
```

```python
class PPOMemory:
    def __init__(self, state_shape, action_shape, T):
        self.state_shape = state_shape
        self.action_shape = action_shape
        self.T = T
        
        self.reset()
        
    def reset(self):
        self.log_prob = np.zeros((self.T, 1), dtype=np.float64)
        self.returns = np.zeros((self.T, 1), dtype=np.float64)
        self.values = np.zeros((self.T, 1), dtype=np.float64)
        
        self.states = np.zeros((self.T, *self.state_shape), dtype=np.float64)
        self.states_ = np.zeros((self.T, *self.state_shape), dtype=np.float64)
        self.actions = np.zeros((self.T, self.action_shape), dtype=np.float64)
        
        self.dones = np.zeros((self.T, 1), dtype=np.bool)
        
        self.mem_count = 0
         
    def store(self, log, r, v, s, s_, a, done):
        idx = min(self.mem_count, self.T)
        self.log_prob[idx] = log
        self.returns[idx] = r
        self.values[idx] = v
        
        self.states[idx] = s
        self.states_[idx] = s_
        self.actions[idx] = a
        
        self.dones[idx] = done
        
        self.mem_count += 1
        
    def sample(self, bs):
        indices = np.random.randint(0, self.mem_count-1, size=(bs))
        
        log = self.log_prob[indices]
        r = self.returns[indices]
        v = self.values[indices]
        
        s = self.states[indices]
        s_= self.states_[indices]
        a = self.actions[indices]

        d = self.dones[indices]
        
        return log, r, v, s, s_, a, d
        
```

```python
class ActorCritic(nn.Module):
    def __init__(self, input_shape, output_shape, std=0.0, lr=0.001):
        super().__init__()
        
        self.input_shape = input_shape
        self.output_shape = output_shape
        
        self.fc = nn.Linear(*input_shape, 128)
        self.fc1 = nn.Linear(128, 256)
        self.fc2 = nn.Linear(256, 128)
        
        self.mu = nn.Linear(128, output_shape)
        self.sigma = nn.Linear(128, 1)
        
        self.log_std = nn.Parameter(torch.ones(output_shape)*std)
        
        self.optimizer = optim.Adam(self.parameters(), lr=lr)
        
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.to(self.device)
        
    def forward(self, x):
        x = torch.relu(self.fc(x))
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        
        mu = self.mu(x)
        sigma = self.sigma(x)
        
        std = self.log_std.exp().expand_as(mu)
        dist = torch.distributions.Normal(mu, std)
        
        return dist, sigma
```

```python
class PPOAgent:
    def __init__(self, env, input_shape, num_actions, 
                 batch_size=5, timesteps=20, epochs=4, 
                 lr=0.001, gamma=0.99, tau=0.95, eps=0.2):
        
        self.input_shape = input_shape
        self.num_actions = num_actions
        self.env = env
        
        self.epsilon = eps
        
        self.batch_size=5
        self.timesteps=20
        self.epochs=4
        self.gamma = gamma
        self.tau = tau
        
        self.net = ActorCritic(self.input_shape, self.num_actions, lr=lr)
        self.memory = PPOMemory(self.input_shape, self.num_actions, self.timesteps)
        
        self.device = self.net.device
        
        self.log_prob = None
        
        self.running_score = None
        
    def choose_action(self, x, dist_=False):
        x = torch.tensor(x).float().to(self.device)
        dist, vals = self.net(x)
        action = dist.sample()
        self.log_prob = dist.log_prob(action)
        
        if dist_:
            return [action, vals, dist]
        else:
            return [action]
    
    def play_env(self, episodes=1):
        rewards = []
        with torch.no_grad():
            for i in range(episodes):
                done = False
                score = 0
                obs = self.env.reset()

                while not done:
                    action = self.choose_action(obs)[0].cpu().numpy()
                    obs_, reward, done, info = self.env.step(action)

                    score += reward
                    obs = obs_
                    
                    self.running_score = score
                rewards.append(score)
                
        return rewards
    
    def compute_advantages(self, rewards, value, obs_, dones):
        advantages = []
        returns = self.compute_returns(rewards)
        for step in range(len(returns)):
            _, nxt = self.net(obs_[step])
            adv = returns[step] + ((self.tau*self.gamma)**step) * nxt * dones[step] - value[step] 
            advantages.append(adv)
        
        return advantages
    
    def compute_returns(self, rewards):
        returns = []
        for step in range(len(rewards)):
            ret = rewards[step] * (self.gamma**step)
            returns.append(ret)
            
        return returns
    
    def play_episodes(self, obs=None):
        if timesteps is None:
            timesteps = self.timesteps
        
        obs = env.reset() if obs == None else obs
        with torch.no_grad():
            for i in range(timesteps):
                x = torch.tensor(obs).float().to(self.device)
                actions, value, dist = self.choose_action(x, dist_=True)
                action = actions.cpu().detach().numpy()
                
                obs_, reward, done, info = env.step(action)
                
                if done:
                    self.play_episodes(timesteps=timesteps - (i+1), obs=obs_)

                self.memory.store(dist.log_prob(actions).cpu(), reward, value.cpu(), obs, obs_, action, done)

                obs = obs_
        
        return obs
                
    def train(self):
        for i in range(self.epochs):
            self.net.optimizer.zero_grad()
            
            old_log_probs, rewards, values, obs, obs_, actions, dones = \
                        self.memory.sample(self.batch_size)
                
            old_log_probs = torch.tensor(old_log_probs).float().to(self.device)
            rewards = torch.tensor(rewards).float().to(self.device)
            values = torch.tensor(values).float().to(self.device)
            obs = torch.tensor(obs).float().to(self.device)
            obs_ = torch.tensor(obs_).float().to(self.device)
            actions = torch.tensor(actions).float().to(self.device)
            dones = torch.tensor(dones).bool().to(self.device)
            
                
            # what we need --> new_log_probs, advantage,  entropy, critic_loss
            dist, value = self.net(obs)
            new_log_probs = dist.log_prob(actions)
            
            advantage = self.compute_advantages(rewards, values, obs_, dones)
            
            entropy = dist.entropy().mean()
            
            returns = torch.tensor(self.compute_returns(rewards)).float().to(self.device)
            
            rt = new_log_probs / old_log_probs
            a_loss = torch.min(rt, torch.clamp(rt, 1-self.epsilon, 1+self.epsilon))
            c_loss = returns - value
            
            loss = (a_loss + 0.5 * c_loss - 0.001 * entropy).mean()
            loss.backward()
            
            self.net.optimizer.step()
            
```

```python
# hyper parameterz
max_frames = 15000
env = gym.make('Pendulum-v0')

agent = PPOAgent(env, env.observation_space.shape, env.action_space.shape[0])

rewards = []
scores = []
avgs = []
frame_idx = 0
done = False



# play episodes 
# if memory is full
# update, clear the memory.

for i in range(500):
    score = 0
    done = False
    obs = env.reset()
    while not done:
        with torch.no_grad():
            x = torch.tensor(obs).float().to(agent.device)
            actions, value, dist = agent.choose_action(x, dist_=True)
            action = actions.cpu().detach().numpy()

            obs_, reward, done, info = env.step(action)
            score += reward

            self.memory.store(dist.log_prob(actions).cpu(), reward, value.cpu(), obs, obs_, action, done)

            obs = obs_
            
        if agent.memory.mem_count == agent.timesteps:
            agent.train()

            agent.memory.reset()

    
    scores.append(score)
    avg = np.mean(scores[-100:])
    
    avgs.append(avg)

    rewards = np.r 







    
    

while frame_idx < max_frames and not done:
    frame_idx +=1 

    agent.play_episodes()
    
    agent.train()
    
    _ = agent.play_env(episodes=10)
    scores.append(agent.running_score)
    avg = np.mean(scores)
    avgs.append(avg)
    print(f'Frame_idx: {frame_idx}, Avg_score: {avg}, Reward: {agent.running_score}')
    
    if frame_idx % 100 == 0:
        reward = agent.play_env(episodes=10)
        for i in reward:
            rewards.append(i)
        plot_fn(rewards, avgs)
        
    agent.memory.reset()
```
